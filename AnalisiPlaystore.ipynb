```python
#In quanto il campo installazioni è una stringa, creo un nuovo campo con il format Int
apps['InstallsInt'] = apps['Installs'].str.replace(',', '').str.extract('(\d+)').astype(int)
```

```python
#Normalizzo il campo Rating
min_rating = apps['Rating'].min()
max_rating = apps['Rating'].max()
apps['RatingNorm'] = (apps['Rating'] - min_rating) / (max_rating - min_rating)
```

```python
#In quanto il campo Reviews è un oggetto, creo un nuovo campo con il format Int
apps['ReviewsInt'] = apps['Reviews'].astype(int)
```

```python
#Normalizzo il campo Reviews
min_reviews = apps['ReviewsInt'].min()
max_reviews = apps['ReviewsInt'].max()
apps['ReviewsIntNorm'] = (apps['ReviewsInt'] - min_reviews) / (max_reviews - min_reviews)
```

```python
#Normalizzo il campo InstallsInt
min_installs = apps['InstallsInt'].min()
max_installs = apps['InstallsInt'].max()
apps['InstallsIntNorm'] = (apps['InstallsInt'] - min_installs) / (max_installs - min_installs)
```

```python
#Definisco la correlazione tra punteggio, recensioni e installazioni
correlation_RatRevInst = apps[['RatingNorm', 'ReviewsIntNorm', 'InstallsIntNorm']].corr()
```

```python
#Mostro la correlazione tra punteggio, recensioni e installazioni
correlation_RatRevInst
```

```python
#Creo un grafico di correlazione tra punteggio, recensioni e installazioni

#Ritengo che a seconda della categoria può variare significativamente
#la fidelizzazione/interazione degli utenti nei confronti delle app
#in base alla popolarità,coinvolgimento,utilità e marketing

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_RatRevInst, annot=True, cmap='coolwarm')
plt.title('Correlation Rating/Reviews/Installs')
plt.show()
```

```python
# Inserisco conteggio delle app per categoria in una variabile
countAppsxCat = apps['Category'].value_counts()
```

```python
#Creo un grafico delle app per categoria utilizzando la variabile creata
#Con questo cerco di capire quali sono le categorie con più competizione
countAppsxCat.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('App Number')
plt.title('Apps per Category')
plt.show()
```

```python
#Inserisco in una variabile la media dei punteggi per categoria
avgRatxCat = apps.groupby('Category')['Rating'].mean()
```

```python
#Creo un grafico della media dei punteggi per categoria  utilizzando la variabile creata
#Sto cercando di capire se in base alla categoria le valutazioni possono dipendere dall'utilità del genere dello strumento
plt.figure(figsize=(12, 6))
plt.plot(avgRatxCat.index, avgRatxCat.values, marker='o')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Average Rating per Category')
plt.xticks(rotation=90)
plt.show()
```

```python
#In quanto il campo installazioni è una stringa, creo un nuovo campo con il format Int
apps['PriceFloat'] = apps['Price'].str.replace('$', '').astype(float)
```

```python
#Creo un nuovo dataset contenente solo le app a pagamento
#Sto isolando le app a pagamento per studiarle separatamente
paid_apps = apps.loc[apps['Type'] == 'Paid']
```

```python
#Normalizzo il campo PriceFloat
min_installs = apps['PriceFloat'].min()
max_installs = apps['PriceFloat'].max()
apps['PriceFloatNorm'] = (apps['PriceFloat'] - min_installs) / (max_installs - min_installs)
```

```python
#Definisco la correlazione tra prezzo e rating
correlation_RatePrice = apps[['RatingNorm', 'PriceFloatNorm']].corr()
```

```python
#Creo un grafico di correlazione tra prezzo e rating
#Sto cercando di capire se c'è effettivamente una relazione inversa tra il punteggio e il prezzo
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_RatePrice, annot=True, cmap='coolwarm')
plt.title('Correlation Rating/Price')
plt.show()
```

```python
#Definisco la correlazione tra prezzo e installazioni
correlation_InstPrice = apps[['InstallsIntNorm', 'PriceFloatNorm']].corr()
```

```python
#Creo un grafico di correlazione tra prezzo e installazioni
#Approfondisco lo studio della relazione inversa sostituendo il rating con le installazioni, 
#notando che la relazione si accentua
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_InstPrice, annot=True, cmap='coolwarm')
plt.title('Correlation Installs/Price')
plt.show()
```

```python
#Incrocio i dati per categoria di prezzo, installazioni, rating e reviews
#Sto cercando di capire le caratteristiche generali per ogni categoria
apps.groupby('Category').agg({
    'PriceFloat': lambda x: round(x.mean(), 2),
    'InstallsInt': lambda x: round(x.mean(), 2).astype(int),
    'Rating': lambda x: round(x.mean(), 2),
    'ReviewsInt': lambda x: round(x.mean(), 2).astype(int)
})
```
